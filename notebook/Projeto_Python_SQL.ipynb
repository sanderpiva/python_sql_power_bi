# 5. Método para definir as informações gerais, dados estatísticos básicos e valores NaN nas tabelas

```python
def info_tabela(df, nome_tabela):

  print("==========================================")
  print()
  print(f"1 Informações gerais: {nome_tabela}")
  print()
  print(df.info())
  print()
  print("==========================================")
  print()
  print(f"2 Primeiros registros: {nome_tabela}")
  print()
  print(df.head())
  print()
  print("==========================================")
  print()
  print(f"3 Estatíticas gerais: {nome_tabela}")
  print()
  print(df.describe())
  print("==========================================")
  print()
  print(f"4 Contagem de registros NaN: {nome_tabela}")
  print()
  print(df.isna().sum())
  print()
  print("==========================================")

```

# 7. Método de consulta envolvendo todas as tabelas da base de dados

```python
def results_query_inner_join(query, conn):
    print()
    df_query = pd.read_sql_query(query, conn)

    print(f"\nResultado da consulta SQL (Prévia):")
    print(df_query.head())
    print()
    print(df_query.info())
    print()
    print("===== Salvando dados como tabela =====")

    return df_query
    print("================================================")
```

# 8. Query para consultar dados de interesse com instruções SQL

```python
query_sql = """
    SELECT
        pd.ano, cl.Pais,
        pd.Prioridade_Pedido, pt.Categoria, pt.SubCategoria,
        COUNT(DISTINCT pd.ID_Pedido) as total_pedidos,
        ROUND(SUM(vd.Custo_Envio), 2) as total_custo_envio,
        ROUND(SUM(vd.Valor_Venda), 2) as total_faturamento
    FROM TB_DSA_CLIENTES as cl
    INNER JOIN TB_DSA_VENDAS as vd ON cl.ID_Cliente = vd.Cliente
    INNER JOIN TB_DSA_PEDIDOS as pd ON vd.Pedido = pd.ID_Pedido
    INNER JOIN TB_DSA_PRODUTOS as pt ON vd.Produto = pt.ID_Produto
    WHERE pd.Prioridade_Pedido IN ('Medio', 'Alto')
    GROUP BY pd.ano, cl.Pais, pd.Prioridade_Pedido, pt.Categoria, pt.SubCategoria
    ORDER BY total_faturamento DESC
"""
```

# 9. Processamento da query de consulta dos dados de interesse

```python
processed_query = results_query_inner_join(query_sql, conn)
```
